```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.05):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
x = np.linspace(-10, 10, 400)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid(x), 'r')
plt.title("Sigmoid function")
plt.grid()
```

```python
plt.subplot(2, 2, 2)
plt.plot(x, tanh(x), 'g')
plt.title("Tanh function")
plt.grid()
```

```python
plt.subplot(2, 2, 3)
plt.plot(x, relu(x), 'b')
plt.title("ReLU function")
plt.grid()
```

```python
plt.subplot(2, 2, 4)
plt.plot(x, leaky_relu(x), 'm')
plt.title("Leaky ReLU function")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
sample_input = np.array([2.0, 1.0, 0.1])
print("Softmax Output for [2.0, 1.0, 0.1]:", softmax(sample_input))
```
